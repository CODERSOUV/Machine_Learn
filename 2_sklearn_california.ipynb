**Linear Regression**

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
```

```python
df=fetch_california_housing()
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset
```

```python
X=dataset
y=df.target
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y
```

```python
y_train
```

```python
#Standarize the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#Inverse Transform
scaler.inverse_transform(X_train)
```

```python
#Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
#importing the model
from sklearn.linear_model import LinearRegression   
#Cross Validation
from sklearn.model_selection import cross_val_score
reg_model=LinearRegression()
reg_model.fit(X_train,y_train)
```

```python
mse=cross_val_score(reg_model,X_train,y_train,cv=5,scoring='neg_mean_absolute_error')
```

```python

np.mean(mse)
```

```python
reg_pred=reg_model.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test)
```

```python
sns.kdeplot(reg_pred-y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(reg_pred,y_test)
```

**Ridge Regression**

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regressor=Ridge()
```

```python
parameters={'alpha':[10,20,30,35,40,45,50,55,100]}
ridgecv=GridSearchCV(ridge_regressor,parameters,scoring='neg_mean_absolute_error',cv=5)
ridgecv.fit(X_train,y_train)
```

```python
ridgecv.best_params_
```

```python
ridgecv.best_score_
```

```python
ridge_pred=ridgecv.predict(X_test)
```

```python
ridge_pred
```

```python
sns.displot(ridge_pred-y_test,kind='kde')
```

**Lasso Regression**

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
```

```python
lasso
```

```python
parameters={'alpha':[10,20,30,35,40,45,50,55,100]}
lassocv=GridSearchCV(lasso,parameters,scoring='neg_mean_absolute_error',cv=5)
lassocv.fit(X_train,y_train)
```

```python
lassocv.best_params_
```

```python
lassocv.best_score_
```

```python
lasso_pred=lassocv.predict(X_test)
```

```python
import seaborn as sns
sns.displot(lasso_pred-y_test,kind='kde')
```
